```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('students_marks.csv')
def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    if marks >= 80 :
        return "A"
    if marks >= 70:
        return "B"
    if marks >= 60:
        return "C"
    else:
        return "D"
df['total']=df[['Math','Science','English']].sum(axis=1)
df['percentage'] = df['total']/3
df['Grade']=df['percentage'].apply(calculate_grade)
highest =df['percentage'].max()
lowest =df['percentage'].min()
avg =df['percentage'].mean()
topper=df.loc[df['percentage'] == highest,"Name"].values[0]
fail_count =df[df['percentage'] < 60].shape[0]
print(f"Highest_percentage:{highest}({topper})")
print(f"lowest_per:{lowest}")
print(f"avg_percentage:{avg:.2f}")
print(f"number of student faill:{fail_count}")
df.to_csv('marks_calculation.csv',index =False)
plt.figure(figsize =(4,4))
plt.bar(df['Name'],df['percentage'],color = 'skyblue')
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Student performeance")
plt.savefig("report")
plt.show()
```

```python
plt.pie(
    df["Grade"].value_counts(),
    labels=df["Grade"].value_counts().index,
    autopct="%1.1f%%"
)
plt.title("StudentPercentage")
plt.show()
```
